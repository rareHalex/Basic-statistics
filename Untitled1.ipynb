```python
label = df.pop('Churn')
X_train, X_test, y_train, y_test = train_test_split(df, label, test_size=0.15)
cat_cols = df.dtypes[df.dtypes == 'object'].keys()
```

```python
init_model = cat.CatBoostClassifier()

init_model.fit(X_train, y_train, 
          eval_set=(X_test, y_test),
          cat_features=cat_cols,plot = True)

print(f'       : {roc_auc_score(y_test, init_model.predict_proba(X_test)[:, 1])}')
```
